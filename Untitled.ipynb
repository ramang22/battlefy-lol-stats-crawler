```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import scipy.stats as s
import datetime
import matplotlib.patches as mpatches
```

```python
def getTime(x):
    minutes = x.split(":")[1]
    seconds = x.split(":")[2]
    if int(seconds) >= 30:
        return int(minutes)+1
    else :
        return int(minutes)
df =  pd.read_excel("./lol_stats.xlsx")
#df['TIME'] =  pd.to_datetime(df['TIME'], format='%H:%M:%S')
df['TIME'] = df['TIME'].apply(lambda x : getTime(x))
```

```python
grp = df.groupby(by=["Meno"]).sum()
```

```python
grp['TEAM'] = ""
```

```python
for index, row in grp.iterrows():
    grp.loc[index,'TEAM'] = df[df['Meno'] == index]['TEAM NAME'].iloc[0]
```

```python
grp = grp[grp['TEAM'].str.contains("ESTUBA")]
```
